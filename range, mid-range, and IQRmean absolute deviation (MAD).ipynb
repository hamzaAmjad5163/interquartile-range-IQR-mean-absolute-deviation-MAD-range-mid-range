```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/task 2.csv')

# Display the first ten rows of the dataset
print("First ten rows of the dataset:")
print(df.head(10))

# Task 2: Data Analysis
variables = ["Temperature (째C)", "Humidity (%)", "Wind Speed (km/h)", "Precipitation (mm)"]

for variable in variables:
    print("\nVariable:", variable)

    data_range = df[variable].max() - df[variable].min()
    print("Range:", data_range)

    mid_range = (df[variable].max() + df[variable].min()) / 2
    print("Mid-Range:", mid_range)

    Q1 = df[variable].quantile(0.25)
    Q3 = df[variable].quantile(0.75)
    IQR = Q3 - Q1
    print("Interquartile Range (IQR):", IQR)

    mad = df[variable].mad()
    print("Mean Absolute Deviation (MAD):", mad)

    # Create box plot with strip plot
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[variable], width=0.5)
    sns.stripplot(x=df[variable], color='orange', jitter=0.4, size=5)
    plt.title(f"Box Plot with Strip Plot of {variable}")
    plt.xlabel(variable)
    plt.show()

plt.figure(figsize=(10, 8))
sns.pairplot(df[variables])
plt.show()

# Task 3: Advanced Analysis
extreme_conditions = df[(df["Temperature (째C)"] > 30) & (df["Humidity (%)"] < 30) |
                        (df["Wind Speed (km/h)"] > 40) | (df["Precipitation (mm)"] > 50)]
print("\nDays with Extreme Weather Conditions:")
print(extreme_conditions)
```
